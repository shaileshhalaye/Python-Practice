# What is Seaborn??



Seaborn is a popular data visualization library for Python.
Seaborn combines aesthetic appeal and technical insights – two crucial cogs in a data science project.
Seaborn is a Python data visualization library based on the Matplotlib library. It provides a high-level interface for drawing attractive and informative statistical graphs.

**Introduction**
<ul>
There is just something extraordinary about a well-designed visualization. The colors stand out, the layers blend nicely together, the contours flow throughout, and the overall package not only has a nice aesthetic quality, but it provides meaningful insights to us as well.
This is quite important in data science where we often work with a lot of messy data. Having the ability to visualize it is critical for a data scientist. Our stakeholders or clients will more often than not rely on visual cues rather than the intricacies of a machine learning model.
There are plenty of excellent Python visualization libraries available, including the built-in matplotlib. But seaborn stands out for us. It combines aesthetic appeal seamlessly with technical insights, as we’ll soon see.
Seaborn’s greatest strengths are its diversity of plotting functions. It allows us to make complicated plots even in a single line of code!.




# Why should you use Seaborn versus matplotlib?
we've been talking about how awesome seaborn is so you might be wondering what all the fuss is about.

I’ll answer that question comprehensively in a practical manner when we generate plots using seaborn. For now, let’s quickly talk about how seaborn feels like it’s a step above matplotlib.

Seaborn makes our charts and plots look engaging and enables some of the common data visualization needs (like mapping color to a variable or using faceting). Basically, it makes the data visualization and exploration easy to conquer. And its serious, that is no easy task in data science.

**If matplotlib “tries to make easy things easy and hard things possible”, seaborn tries to make a well-defined set of hard things easy too.**

There are essentially a couple of (big) limitations in matplotlib that Seaborn fixes:

Seaborn comes with a large number of high-level interfaces and customized themes that matplotlib lacks as it’s not easy to figure out the settings that make plots attractive
Matplotlib functions don’t work well with dataframes, whereas seaborn does

# Plot a Scatter Plot between Total_bill and tip

```python
#plot here
sns.relplot(x='total_bill',y='tip',data=data,kind='scatter',style='smoker',hue='sex' ,size='size')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

 **Plot using day and tip**

 note=jitter

```python
sns.catplot(x='day',y='tip',data=tips,kind='strip')##jitter=2)#jitter makes noise ,bydefault value is 0.1
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='swarm')
```

```python
#plot using strip plot directly

sns.stripplot(x='day',y='tip',data=tips)
```

```python
#plot using swarm plot directly

sns.swarmplot(x='day',y='tip',data=tips)
```

# **2.Categorical Distribution Plots**

```python
#1.Box Plot

#Univariant -- means plotting on only one column    ----- 25%-75 is known as inter quartile range(IQR)
sns.boxplot(tips["tip"])  ##diamonds are outliers means 50 students class example
#lines are known as Whiskers
```

```python
#multivariant -- means plotting on more than one column

sns.catplot(x='day',y='tip',data=tips,kind = 'box')
```

```python
sns.catplot(x='day',y='tip',data=tips,kind = 'box',hue='sex')
```

# **Violin Plot**

```python
#combination on box plot and distribution of data  ##Mirror Image
```

```python
#Axes Level
sns.violinplot(tips["tip"]) #output white dot is median(50%),whole is box plot
```

```python
#figure Level
sns.catplot(x='day',y='tip',data=tips,kind = 'violin',hue='sex',split=True)
```

# Pointplot, Barplot & Countplot

```python
sns.pointplot(tips["tip"])
```

```python
sns.catplot(x='day',y='tip',data=tips,kind = 'point')
```

```python
#numerical data to categorical data
sns.barplot(tips["tip"])
```

```python
sns.catplot(x='day',y='tip',data=tips,kind = 'bar',estimator=np.median)#hue='sex')## Black bar is called as error bar which
##gives mean of specific day #this gives statistics results such as mean ,median,std.
##by default is mean(##estimator=np.median())
```

```python
sns.catplot(x='time',data=tips,kind = 'count',hue='sex')
```

# Heatmaps and Clustermaps

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc=sum)
```

```python
plt.figure(figsize=(10,5))#heatmap is mostly used in machine learning
sns.heatmap((x),annot=True,fmt='d') ##d means result in integer
```

```python
sns.clustermap(x)
```
