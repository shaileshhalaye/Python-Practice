```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Line plot
x=np.arange(1,11)
x
```

```python
y=x*2
y
```

```python
#Adding Title and Labels
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-lable")
plt.ylabel("y-lable")
plt.show()
```

```python
#Changing Line Aesthetic
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.title('x vs y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show
```

```python
#Adding two lines in the same plot
x=np.arange(1,11)
y1=x*2
y2=x*3
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show
```

```python
plt.plot(x,y1,color='g',linestyle='-.')
plt.plot(x,y2,color='y',linewidth=2)
plt.title("Line Plot")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.grid(True)
plt.show()
```

```python
#Adding sub-plots
x=np.arange(1,11)
y1=x*2
y2=x*3
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linestyle='-.',linewidth=2)
plt.title("x vs y1")
plt.xlabel("This is x")
plt.ylabel("This is y")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle=':',linewidth=3)
plt.title("x vs y2")
plt.xlabel("This is x")
plt.ylabel("This is y2")
plt.grid(True)

plt.show()
```

```python
#Bar Plot
student={'Bob':10,'Matt':56,'Sam':27}
```

```python
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Horizontal Bar Plot
plt.barh(names,values,color='g')
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Creating basic scatter-plot
x=[10,20,30,40,50,60,70,80]
y=[8,1,7,0,3,7,3,2]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,marker="*",c="g",s=100)
plt.show()
```

```python
#Adding two markers in the same plot
x=[10,20,30,40,50,60,70,80,90]
y=[8,1,7,2,0,3,7,3,2]
z=[7,2,5,6,9,1,4,5,3]
plt.scatter(x,y,marker="*",c="r",s=100)
plt.scatter(x,z,marker=".",c="g",s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,z,marker=".",c="r",s=200)
plt.show()
```
