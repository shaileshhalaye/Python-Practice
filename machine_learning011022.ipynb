```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
iris=snr.load_dataset('iris')
```

```python
data=iris.drop(['species'],axis=1)
```

```python
#Histogram
data['petal_length'].diff().plot(kind='hist',bins=200) #bins represent the width of each histogram block
```

```python
data['sepal_width'].diff().plot(kind='hist',bins=200)
```

```python
data['sepal_length'].diff().plot(kind='hist',bins=150)
```

```python
data['petal_width'].diff().plot(kind='hist',bins=150)
```

```python
#bar graph
data.plot(kind='bar')
```

```python
data.iloc[0:5].plot(kind='bar')
```

```python
data.iloc[0].plot(kind='bar')
```

```python
#pie chart
data.iloc[0].plot.pie(figsize=(5,5),fontsize=20,autopct='%.2f') #value inside the pie upto 2 decimal points %.2f
```

```python
data.iloc[0:3].plot.pie(figsize=(10,10),subplots=True,fontsize=15,autopct='%.2f')
```

```python
#scatter plot
data.plot.scatter(x='sepal_length',y='petal_length')  #used to compare any two parameters in the dataset..for 3 max
```

```python
data.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width') #3 parameters showing
```

```python
data.plot.scatter(x='sepal_length',y='petal_length',c='r')  #'r' means red, 'g' means green
```

```python
tips=snr.load_dataset('tips')
```

```python
tips.head()
```

```python
#relational plot
#the no. of featurers can be compare 4,5,6 (more than 3 features)
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time',col='sex')
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='time')
```

```python
snr.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time',col='sex',size='size')
```

```python
#catplot
snr.catplot(x='total_bill',y='day',data=tips)
```

```python
#boxplot
snr.catplot(x='day',y='tip',hue='sex',kind='box',data=tips)
```
